```python
import numpy as np
import pandas as pd
import pickle
import numpy as np
```

```python
# Read the cleaned dataframes
names = ["biology","cooking","crypto","diy","robotics","travel"]

all_dfs = [pd.read_pickle("cleaned_trim/"+names[i]+"_trim_clean.dat") for i in range(len(names))]
```

```python
def ohe(df):
    # Split the tags into separate columns
    dfT = df['tags'].str.split(" ", expand = True)
    
    # Get the unique tags
    tags = list(set(" ".join(df['tags']).split(" ")))
    
    # Create new df where columns are unique tags
    newDF = pd.DataFrame(columns = tags)
    
    # One hot encoding
    ohe = dfT.stack().str.get_dummies().sum(level=0)
    
    # Add the title+text to one hot encoding df
    dat_ohc = pd.concat([df['text'],ohe], axis = 1, sort = False)
    
    return dat_ohc
```

```python
# Get ohe for all catagories. Each ohe df is stored as an element of a list.
ohe_dfs = [ohe(all_dfs[i]) for i in range(len(all_dfs))]

names = ["biology","cooking","crypto","diy","robotics","travel"]
for i in range(len(names)):
    ohe_dfs[i].to_pickle("ohe_"+names[i]+".dat")
```

```python
# Get ohe for entire dataset
combined_dfs = pd.concat(all_dfs)
ohe_combined = ohe(combined_dfs)
ohe_combined.to_pickle("ohe_all.dat")
```
