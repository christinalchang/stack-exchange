```python
import pandas as pd
from plotnine import *
```

```python
biology = pd.read_pickle("/Users/Troy/Desktop/DataData/biology_finalDF")
cooking = pd.read_pickle("/Users/Troy/Desktop/DataData/cooking_finalDF")
crypto = pd.read_pickle("/Users/Troy/Desktop/DataData/crypto_finalDF")
diy = pd.read_pickle("/Users/Troy/Desktop/DataData/diy_finalDF")
robotics = pd.read_pickle("/Users/Troy/Desktop/DataData/robotics_finalDF")
travel = pd.read_pickle("/Users/Troy/Desktop/DataData/travel_finalDF")
full = pd.read_pickle("/Users/Troy/Desktop/DataData/full_finalDF")
```

```python
len(biology.columns)
```

```python
y = [(len(biology.columns)-2), (len(cooking.columns)-2), 
     (len(crypto.columns)-2), (len(diy.columns)-2), 
     (len(robotics.columns)-2), (len(travel.columns)-2),
    (len(full.columns)-2)]
x = ["Biology", "Cooking", "Cryptography", "D.I.Y", "Robotics", "Travel",
    "Full"]
paraDF = pd.DataFrame({"Names" : x, "Values": y})
```

```python
(
ggplot(paraDF, aes(x = "Names", y = "Values")) 
    + geom_bar(stat = "identity", color = "darkorange")
    + theme(axis_text_x  = element_text(angle = 40, hjust = 1))
    + xlab("Categories")
    + ylab("Number of Tags")
    + ggtitle("Number of Tags per Category after Subset")
)
```
